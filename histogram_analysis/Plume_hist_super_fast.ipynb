```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob


# plt.rc("figure", figsize=(8, 6))
# plt.rc("figure", dpi=100)
```

```python
# parameters used for the simulation
r_tot = 70774597466.3827
r_bot = 0.4 * r_tot
r_top = 0.9 * r_tot
Nr = 504
Ntheta = 360
Nphi = 720
dr_grid = (r_top - r_bot) / Nr
r_conv = 5.149819426617924E+10
```

```python
def read_file(fname):
    if int(fname[-4:]) % 50 == 0:
        print(fname[-4:])
    with open(fname, 'r') as f:
        t = float(f.readline())
        df = pd.read_csv(f, sep='\s+',
                    #names=["num_theta", "theta_deg", "HP1", "r1", "renv", "HP2", "r21", "rcore"])
        #names=["num_theta", "theta_deg", "HP1", "r1", "renv", "HP2", "r21", "rcore", "renvdown", "renvup", "rcoredown", "rcoreup"])
        names=["num_phi", "phi_deg", "num_theta", "theta_deg", "HP1", "r1", "r2","r3","r4","r5", "HP2", "r21", "r22","r23","r24","r25"],
        usecols=["num_phi", "phi_deg", "num_theta", "theta_deg", "HP1", "r1", "r2"])
        df["time"] = t
        return df
# r21=zerok (Kin. flux)  
# r22=zeroc (heat flux fluxcr using horizontal average T(i,t))     r23=zeroc1 (heat flux fluxcr1  usint T init)
# r24=zeroc2  (heat flux fluxcr2 using time/horizontal average <T(i)>=avtar(i))  
# r25=zeroc3 (heat flux fluxcr3 using time average of T(i,j) <T(i,j)>=avt(i,j))
```

#### Analysis

```python
df_Om0['depth_kin'] =  - (df_Om0["r1"] - r_conv) / r_tot
df_Om1['depth_kin'] =  - (df_Om1["r1"] - r_conv) / r_tot
df_Om5['depth_kin'] =  - (df_Om5["r1"] - r_conv) / r_tot

df_Om0['depth_enth'] =  - (df_Om0["r2"] - r_conv) / r_tot
df_Om1['depth_enth'] =  - (df_Om1["r2"] - r_conv) / r_tot
df_Om5['depth_enth'] =  - (df_Om5["r2"] - r_conv) / r_tot

df_Om0
```

```python
df_Om0.set_index('num_phi', append=True, inplace=True)
```

```python
df_Om1.set_index('num_phi', append=True, inplace=True)
df_Om5.set_index('num_phi', append=True, inplace=True)
```

```python
Om0_maxes = df_Om0.groupby('time')['depth_kin'].idxmax()
Om1_maxes = df_Om1.groupby('time')['depth_kin'].idxmax()
Om5_maxes = df_Om5.groupby('time')['depth_kin'].idxmax()
```

```python
Om0_thetas = [t[1] for t in Om0_maxes]   # list of theta value
Om0_phis = [t[2] for t in Om0_maxes]

Om1_thetas = [t[1] for t in Om1_maxes]   # list of theta value
Om1_phis = [t[2] for t in Om1_maxes]

Om5_thetas = [t[1] for t in Om5_maxes]   # list of theta value
Om5_phis = [t[2] for t in Om5_maxes]
```
