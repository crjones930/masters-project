```python
import matplotlib.pyplot as plt
import numpy as np
import glob
import pandas as pd
```

```python
PATTERN= "/z2/users/cj517/lscaledir_3D/r_min0.4_r_max0.9_from2d_v0_Om0/struc*"
all_files = sorted(glob.glob(PATTERN))
all_files
```

```python
spec_file = all_files[0]
```

```python
times = []
for spec_file in all_files:
    with open(spec_file, "r") as f:
        f.readline()
        times.append(float(f.readline().split()[0]))
        f.close()
```

```python
plt.plot(np.diff(times), ls='', marker='o')
```

```python
total_time = times[-1] - times[0]
```

```python
conv = []
for spec_file in all_files:
    with open(spec_file, "r") as f:
        f.readline()
        # print(f.readline().split())
        conv.append(float(f.readline().split()[4]))
        f.close()
```

```python
plt.plot(conv, ls='', marker='.')
```

```python
conv_av = np.mean(conv)
```

```python
total_time / conv_av
```
