```python
def read_file(fname):
    with open(fname, "r") as f:
        f.readline()
        time = f.readline().split()[0]
    df = pd.DataFrame(np.genfromtxt(fname,skip_header=3),columns=headers)
    # df['time'] = time
    return df
```
