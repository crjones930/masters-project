```python
# import modules
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 2D base case

```python

# absolute_path = "/home/cj517/z1_data/r_min0.1_r_max0.97_2d/base/params.nml"
absolute_path = "/home/cj517/my_z2/lscaledir_young_sun/r_min0.1_r_max0.97_2d/time_series.dat"
PATTERN = os.path.relpath(absolute_path)

all_files = sorted(glob.glob(PATTERN))
all_files
```

```python
with open(all_files[0], 'r') as f:
    headers = f.readline().split()
```

```python
data = np.genfromtxt(all_files[0],skip_header=1)
```

```python
df = pd.DataFrame(data)
df.columns = headers
```

```python
df = df.sort_values(by='time')
```

```python
time = df['time']
log_ke = df['log_e_kin_tot']
```

### Plotting

```python
fig,ax = plt.subplots()
ax.plot(time, log_ke, label=r'$0$')
ax.plot(time_Om1, log_ke_Om1, label=r'$1$')
ax.plot(time_Om5, log_ke_Om5, label=r'$5$')

ax.set_ylim(40,43)
ax.set_xlim(0,0.6e8)

ax.minorticks_on()

ax.grid(which='minor', color='gray', linestyle='-', linewidth=0.2)
ax.grid(which='major', color='gray', linestyle='-', linewidth=0.5)

ax.legend(title = r'$\Omega/\Omega_\odot$')

ax.set_ylabel('log_total_ke')
ax.set_xlabel('time/s')


fig.savefig('total_ke.png')
```
